```python
import dotenv
import os

dotenv.load_dotenv()

import nekt

nekt.data_access_token = os.getenv("NEKT_TOKEN")
```

```python
spark = nekt.get_spark_session()
```

```python

points_clientes_df = nekt.load_table(layer_name="Bronze", table_name="points_clientes")
points_clientes_df.createOrReplaceTempView("points_clientes")
```

```python
spark.sql("SELECT * FROM points_clientes LIMIT 10").toPandas()
```
